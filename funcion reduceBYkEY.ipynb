```scala
val sc = spark.sparkContext
```

```scala
val rdd = sc.parallelize(Seq(
  ("casa", 2),
  ("parque", 1),
  ("que", 5),
  ("casa", 1),
  ("escuela", 2),
  ("casa", 1),
  ("que", 1)
))
```

```scala
rdd.collect(
)
```

```scala
val rddReducido = rdd.reduceByKey(_ + _)
rddReducido.collect()
```

```scala
val rddReducido2 = rdd.reduceByKey((x,y) => x + y)
rddReducido2.collect()
```
