```scala
val sc = spark.sparkContext
```

```scala
val rdd = sc.parallelize(Seq(1,2,3,4,5,6,7,8,9))
```

```scala
val rddCuadrado = rdd.map(x => List(x, x * x))
rddCuadrado.collect()
```

```scala
val rddCuadratoFlat = rdd.flatMap(x => List(x, x * x))
rddCuadratoFlat.collect()
```

```scala
val rddTexto = sc.parallelize(Seq("azul rojo verde" , "morado amarillo negro"))
```

```scala
val rddColoroesFlat = rddTexto.flatMap(_.split(" ")) 
rddColoroesFlat.collect()
```
